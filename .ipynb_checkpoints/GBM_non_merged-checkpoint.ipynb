```python
import numpy as np
import pandas as pd
import os
from sklearn.metrics import roc_curve, auc, precision_recall_curve, average_precision_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
import shap
from sklearn import preprocessing
from sklearn.utils import resample
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import precision_score, recall_score, f1_score
from copy import deepcopy
from math import sqrt
from sklearn.ensemble import GradientBoostingClassifier
```

```python
derivation_cohort = pd.read_csv("Derivation_dataset.csv")
```

```python
testing_cohort = pd.read_csv("processed_ecgs_all_labelled.csv")
```

```python
y = derivation_cohort['Outcome_Occlusion_MI']
X = derivation_cohort.drop('Outcome_Occlusion_MI', axis=1)
X_train, X_val, y_train, y_val = train_test_split(X, y, test_size=0.2, stratify=y, random_state=99)
y_test = testing_cohort['OMI']
X_test = testing_cohort.drop(['OMI', 'patient_1'], axis=1)

# **(1) Remove StandardScaler and let GBM handle the raw data**
# **(2) Remove manual missing value imputation (LightGBM and XGBoost handle missing values internally)**
for el in X_train.columns:
    X_train[el].fillna(X_train[el].mode()[0], inplace=True)
    X_val[el].fillna(X_train[el].mode()[0], inplace=True)
    X_test[el].fillna(X_train[el].mode()[0], inplace=True)

y_train = np.ravel(y_train.values)
y_val = np.ravel(y_val.values)
X_test = np.array(X_test.values)

# Keep as DataFrame (GBM models handle categorical features better with DataFrame)
columns = X_train.columns
```

```python
clf = GradientBoostingClassifier(
    n_estimators=109,
    learning_rate=0.1,  # Default is 0.1, can be tuned
    max_depth=3,  # Typical default value, adjust if needed
    min_samples_split=10,  # Adjusted based on RF min_samples_split
    min_samples_leaf=5,  # Adjusted based on RF min_samples_leaf
    subsample=0.75,  # Equivalent to max_samples=0.75
    random_state=42
)

# Optional calibration
clf = CalibratedClassifierCV(clf, cv=5, method="isotonic")
```

```python
def plot_ROC_curve(fpr, tpr, roc_auc, title):
    plt.figure(figsize=[5, 5])
    plt.title(title)
    plt.plot(fpr, tpr, color='b',
                 label=r'GBM (AUC = %0.3f)' % (roc_auc),
                 lw=2, alpha=.8)
    plt.plot([0, 1], [0, 1], linestyle='--', lw=1, color='grey', alpha=.5)
    plt.xlim([-0.05, 1.05])
    plt.ylim([-0.05, 1.05])
    plt.xlabel('False Positive Rate = 1 - Specificity')
    plt.ylabel('True Positive Rate = Sensitivity')
    plt.legend(loc="lower right")
    plt.show()
```

```python
def results(pred_probas):
    OMI_score = np.round(pred_probas[:, 1]*100, decimals=2)
    y_pred = np.where(OMI_score<5, 'Low risk', 'Intermediate risk')
    y_pred = np.where(OMI_score>=20, 'High risk', y_pred)
    
    results = {
        'Low risk': [np.count_nonzero(y_pred == 'Low risk')],
        'Intermediate risk': [np.count_nonzero(y_pred == 'Intermediate risk')],
        'High risk': [np.count_nonzero(y_pred == 'High risk')]
    }
    return(pd.DataFrame(data=results, index=['Count (N)']))
```

```python
def bootstrap_metric_ci(y_true, y_pred, metric_fn, n_bootstrap=1000, alpha=0.05):
    """
    Compute 95% confidence interval using bootstrapping for any metric.
    
    Parameters:
    - y_true: Ground truth labels (0 or 1).
    - y_pred: Predicted values (either probabilities or binary predictions).
    - metric_fn: Metric function (roc_auc_score, recall_score, f1_score).
    - n_bootstrap: Number of bootstrap resamples.
    - alpha: Significance level (default 0.05 for 95% CI).

    Returns:
    - Lower and upper bound of the confidence interval.
    """
    scores = []
    
    for _ in range(n_bootstrap):
        # Resample dataset with replacement
        indices = resample(range(len(y_true)), replace=True, n_samples=len(y_true))
        y_true_bootstrap = np.array(y_true)[indices]
        y_pred_bootstrap = np.array(y_pred)[indices]

        # Special case for AUROC: Use roc_curve and auc()
        if metric_fn == "auroc":
            fpr, tpr, _ = roc_curve(y_true_bootstrap, y_pred_bootstrap)
            score = auc(fpr, tpr)
        else:
            score = metric_fn(y_true_bootstrap, y_pred_bootstrap)

        scores.append(score)
    
    # Compute confidence interval
    lower_bound = np.percentile(scores, 100 * (alpha / 2))
    upper_bound = np.percentile(scores, 100 * (1 - alpha / 2))
    
    return lower_bound, upper_bound
```

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.calibration import CalibratedClassifierCV
from sklearn.metrics import recall_score
import numpy as np

# Define the parameter grid for GBM
param_grid = {
    'n_estimators': [75],
    'learning_rate': [0.2],
    'max_depth': [3],
    'min_samples_split': [5],
    'min_samples_leaf': [ 5],
    'subsample': [0.75]
}

print("Hyperparameter Tuning for GBM on Validation Set")
print("Searching for optimal hyperparameters...")

best_score = 0
best_params = {}
best_model = None
best_calibrated_model = None

# Track progress
total_combinations = (len(param_grid['n_estimators']) * 
                      len(param_grid['learning_rate']) * 
                      len(param_grid['max_depth']) * 
                      len(param_grid['min_samples_split']) * 
                      len(param_grid['min_samples_leaf']) * 
                      len(param_grid['subsample']))
                      
print(f"Total parameter combinations to try: {total_combinations}")
current_combination = 0

# Loop through all parameters
for n_estimators in param_grid['n_estimators']:
    for learning_rate in param_grid['learning_rate']:
        for max_depth in param_grid['max_depth']:
            for min_samples_split in param_grid['min_samples_split']:
                for min_samples_leaf in param_grid['min_samples_leaf']:
                    for subsample in param_grid['subsample']:
                        current_combination += 1
                        if current_combination % 10 == 0:
                            print(f"Progress: {current_combination}/{total_combinations} combinations ({(current_combination/total_combinations*100):.1f}%)")
                        
                        try:
                            # Create base GBM model with current parameters
                            base_gbm = GradientBoostingClassifier(
                                n_estimators=n_estimators,
                                learning_rate=learning_rate,
                                max_depth=max_depth,
                                min_samples_split=min_samples_split,
                                min_samples_leaf=min_samples_leaf,
                                subsample=subsample,
                                random_state=42
                            )
                            
                            # Train on training set
                            base_gbm.fit(X_train, y_train)
                            
                            # First evaluate the base model
                            try:
                                val_probas = base_gbm.predict_proba(X_val)
                                if np.isnan(val_probas).any():
                                    continue
                                    
                                y_pred_val = (val_probas[:,1] >= 0.20).astype(int)
                                recall_val = recall_score(y_val, y_pred_val)
                                
                                # Now try with calibration
                                calibrated_gbm = CalibratedClassifierCV(
                                    base_gbm, 
                                    cv='prefit',  # Use prefit since we already trained the base model
                                    method="isotonic"
                                )
                                calibrated_gbm.fit(X_val, y_val)  # Calibrate on validation set
                                
                                # Evaluate the calibrated model
                                cal_val_probas = calibrated_gbm.predict_proba(X_val)
                                cal_y_pred_val = (cal_val_probas[:,1] >= 0.20).astype(int)
                                cal_recall_val = recall_score(y_val, cal_y_pred_val)
                                
                                # Determine which model (base or calibrated) performed better
                                if cal_recall_val > recall_val:
                                    current_recall = cal_recall_val
                                    use_calibration = True
                                else:
                                    current_recall = recall_val
                                    use_calibration = False
                                
                                print(f"Parameters: n_est={n_estimators}, lr={learning_rate}, depth={max_depth}, "
                                      f"min_split={min_samples_split}, min_leaf={min_samples_leaf}, subsample={subsample}")
                                print(f"Base recall={recall_val:.4f}, Calibrated recall={cal_recall_val:.4f}")
                                
                                # Track best model
                                if current_recall > best_score:
                                    best_score = current_recall
                                    best_params = {
                                        'n_estimators': n_estimators,
                                        'learning_rate': learning_rate,
                                        'max_depth': max_depth,
                                        'min_samples_split': min_samples_split,
                                        'min_samples_leaf': min_samples_leaf,
                                        'subsample': subsample,
                                        'use_calibration': use_calibration
                                    }
                                    best_model = base_gbm
                                    if use_calibration:
                                        best_calibrated_model = calibrated_gbm
                            except Exception as e:
                                print(f"Error evaluating model: {e}")
                        except Exception as e:
                            print(f"Error creating/training model: {e}")

# Print final results summary
if best_model is not None:
    print("\n" + "="*50)
    print(f"Best Parameters: {best_params}")
    print(f"Best Validation recall score: {best_score:.4f}")
    
    # Create the final model using the best parameters
    if best_params.get('use_calibration', False) and best_calibrated_model is not None:
        print("Using calibrated model with isotonic calibration")
        clf = best_calibrated_model
    else:
        print("Using base GBM model without calibration")
        base_gbm = GradientBoostingClassifier(
            n_estimators=best_params['n_estimators'],
            learning_rate=best_params['learning_rate'],
            max_depth=best_params['max_depth'],
            min_samples_split=best_params['min_samples_split'],
            min_samples_leaf=best_params['min_samples_leaf'],
            subsample=best_params['subsample'],
            random_state=42
        )
        base_gbm.fit(X_train, y_train)
        clf = base_gbm
else:
    print("\nNo valid model found. Check your data and evaluation criteria.")
```

```python
# Predict on training data
pred_probas_train = clf.predict_proba(X_train)
risk_groups_train = results(pred_probas_train)

# Compute ROC curve and area under the curve
fpr, tpr, thresholds = roc_curve(y_train, pred_probas_train[:,1])
roc_auc = auc(fpr, tpr)
plot_ROC_curve(fpr, tpr, roc_auc, 'Training curve: Derivation cohort (ROC curves)')

# Clinical threshold evaluation
y_pred_clinical = (pred_probas_train[:,1] >= 0.20).astype(int)
recall_clinical = recall_score(y_train, y_pred_clinical)
f1 = f1_score(y_train, y_pred_clinical)

# Compute AUROC CI
auroc_ci_lower_train, auroc_ci_upper_train = bootstrap_metric_ci(y_train, pred_probas_train[:,1], "auroc")

# Compute Recall CI
recall_ci_lower_train, recall_ci_upper_train = bootstrap_metric_ci(y_train, y_pred_clinical, recall_score)

# Compute F1-score CI
f1_ci_lower_train, f1_ci_upper_train = bootstrap_metric_ci(y_train, y_pred_clinical, f1_score)

print(f"Training Recall: {recall_clinical:.3f} (95% CI: {recall_ci_lower_train:.3f}, {recall_ci_upper_train:.3f})")

print(f"Training F1-score: {f1:.3f} (95% CI: {f1_ci_lower_train:.3f}, {f1_ci_upper_train:.3f})")

print(f"Training AUROC: {roc_auc:.3f} (95% CI: {auroc_ci_lower_train:.3f}, {auroc_ci_upper_train:.3f})")
```

```python
# Evaluate on validation set
print("\nVal")
pred_probas_val = clf.predict_proba(X_val)
risk_groups_val = results(pred_probas_val)

# Compute ROC curve and area under the curve for validation set
fpr_val, tpr_val, thresholds_val = roc_curve(y_val, pred_probas_val[:,1])
roc_auc_val = auc(fpr_val, tpr_val)
plot_ROC_curve(fpr_val, tpr_val, roc_auc_val, 'Validation curve: Derivation cohort (ROC curves)')

# Clinical threshold evaluation on val set
y_pred_val_clinical = (pred_probas_val[:,1] >= 0.20).astype(int)
recall_val_clinical = recall_score(y_val, y_pred_val_clinical)
f1_val = f1_score(y_val, y_pred_val_clinical)

# Compute AUROC CI
auroc_ci_lower_val, auroc_ci_upper_val = bootstrap_metric_ci(y_val, pred_probas_val[:,1], "auroc")

# Compute Recall CI
recall_ci_lower_val, recall_ci_upper_val = bootstrap_metric_ci(y_val, y_pred_val_clinical, recall_score)

# Compute F1-score CI
f1_ci_lower_val, f1_ci_upper_val = bootstrap_metric_ci(y_val, y_pred_val_clinical, f1_score)

print(f"Val Recall: {recall_val_clinical:.3f} (95% CI: {recall_ci_lower_val:.3f}, {recall_ci_upper_val:.3f})")

print(f"Val F1-score: {f1_val:.3f} (95% CI: {f1_ci_lower_val:.3f}, {f1_ci_upper_val:.3f})")

print(f"Val AUROC: {roc_auc_val:.3f} (95% CI: {auroc_ci_lower_val:.3f}, {auroc_ci_upper_val:.3f})")
```

```python
# Evaluate on test set
print("\nTest")
pred_probas_test = clf.predict_proba(X_test)
risk_groups_test = results(pred_probas_test)

# Compute ROC curve and area under the curve for validation set
fpr_test, tpr_test, thresholds_test = roc_curve(y_test, pred_probas_test[:,1])
roc_auc_test = auc(fpr_test, tpr_test)
plot_ROC_curve(fpr_test, tpr_test, roc_auc_test, 'Test curve: Derivation cohort (ROC curves)')

# Clinical threshold evaluation on val set
y_pred_test_clinical = (pred_probas_test[:,1] >= 0.20).astype(int)
recall_test_clinical = recall_score(y_test, y_pred_test_clinical)
f1_test = f1_score(y_test, y_pred_test_clinical)

# Compute AUROC CI
auroc_ci_lower_test, auroc_ci_upper_test = bootstrap_metric_ci(y_test, pred_probas_test[:,1], "auroc")

# Compute Recall CI
recall_ci_lower_test, recall_ci_upper_test = bootstrap_metric_ci(y_test, y_pred_test_clinical, recall_score)

# Compute F1-score CI
f1_ci_lower_test, f1_ci_upper_test = bootstrap_metric_ci(y_test, y_pred_test_clinical, f1_score)

print(f"Test Recall: {recall_test_clinical:.3f} (95% CI: {recall_ci_lower_test:.3f}, {recall_ci_upper_test:.3f})")

print(f"Test F1-score: {f1_test:.3f} (95% CI: {f1_ci_lower_test:.3f}, {f1_ci_upper_test:.3f})")

print(f"Test AUROC: {roc_auc_test:.3f} (95% CI: {auroc_ci_lower_test:.3f}, {auroc_ci_upper_test:.3f})")
```
